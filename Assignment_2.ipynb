#2. Data Cleaning and Preprocessing

2.1 Remove the first column or “Column 1” from the dataset. Treat the ‘year’ variable as nominal.

```python
import pandas as pd
data = pd.read_csv("Assignment 2 dataset.csv")
data = data.iloc[:, 1:]

# Convert year to (nominal)
data["year"] = data["year"].astype(str)
print(data.head())
print("\nYear data type:", data["year"].dtype)
```

2.2 Check for duplicate values and remove them.

```python


# Count duplicates
duplicate_count = data.duplicated().sum()
print("Number of duplicate rows before removal:", duplicate_count)

# Remove duplicates
data = data.drop_duplicates()

# Confirm removal
print("Number of duplicate rows after removal:", data.duplicated().sum())
print("Total rows after cleaning:", len(data))
```

2.3 Check for missing values. If a data record (row) only has a few missing values, replace the missing values with the median of the column feature in that specific “Region” variable. If most column values in a data record are missing, remove the data record.

```python
# Check missing values
print("Missing values before cleaning:")
print(data.isnull().sum())


threshold = len(data.columns) / 2
data = data.dropna(thresh=threshold)



numeric_columns = data.select_dtypes(include="number").columns

for col in numeric_columns:
    data[col] = data.groupby("region")[col].transform(
        lambda x: x.fillna(x.median())
    )



print("\nMissing values after cleaning:")
print(data.isnull().sum())

print("\nFinal dataset shape:", data.shape)
```

#3. Exploratory Data Analysis (EDA)


3.1 Describe the variables
- Describe all variables in the dataset.
- For continuous variables: report **range (min, max), mean, median, and distribution**.
- For categorical variables: list unique values.

```python


numeric_columns = data.select_dtypes(include="number").columns
categorical_columns = data.select_dtypes(exclude="number").columns



print("Continuous Variables Summary:\n")

for col in numeric_columns:
    print("Variable:", col)
    print("Min:", data[col].min())
    print("Max:", data[col].max())
    print("Mean:", data[col].mean())
    print("Median:", data[col].median())
    print("-" * 40)



print("\nCategorical Variables:\n")

for col in categorical_columns:
    print("Variable:", col)
    print("Unique Values:", data[col].unique())
    print("Number of Unique Values:", data[col].nunique())
    print("-" * 40)
```

3.2 Inspect the earliest recorded date
- Find the earliest `Date`.
- Check if there are avocado prices recorded from the earliest date up to 2010.
- Comment: does the earliest data point look reasonable? Keep or remove?

```python
data.columns = data.columns.str.strip().str.lower()

data["date"] = pd.to_datetime(data["date"])

earliest_date = data["date"].min()
print("Earliest date in dataset:", earliest_date)

before_2010 = data[data["date"].dt.year <= 2010]

print("Number of records from earliest date up to 2010:", len(before_2010))

print(before_2010.head())
```

3.3 Highest average price
- Find the highest value in "AveragePrice".
- Report which region it belongs to.
- Describe how you obtained the result.

```python

data.columns = data.columns.str.strip().str.lower()

max_price = data["averageprice"].max()

highest_row = data[data["averageprice"] == max_price]

print("Highest Average Price:", max_price)
print("\nRegion(s) with Highest Average Price:")
print(highest_row["region"])
```
